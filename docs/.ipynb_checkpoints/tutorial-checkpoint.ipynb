### 2. Importing Cabana

### 3. Setting Up Parameters
Cabana requires a YAML parameter file to configure the analysis. Below is an example of creating such a file.

```python
import yaml
params = {
    "Configs": {
        "Segmentation": True,
        "Quantification": True,
        "Gap Analysis": True
    },
    "Segmentation": {
        "Number of Labels": 32,
        "Max Iterations": 32,
        "Normalized Hue Value": 0.95,
        "Color Threshold": 0.2,
        "Max Size": 2048,
        "Min Size": 64
    },
    "Detection": {
        "Dark Line": True,
        "Extend Line": False,
        "Min Line Width": 7,
        "Max Line Width": 15,
        "Line Width Step": 2,
        "Low Contrast": 100,
        "High Contrast": 200,
        "Minimum Line Length": 10,
        "Maximum Line Length": 0
    },
    "Quantification": {
        "Minimum Branch Length": 5,
        "Minimum Curvature Window": 5,
        "Maximum Curvature Window": 15,
        "Curvature Window Step": 5,
        "Maximum Display HDM": 230,
        "Contrast Enhancement": 0.1
    },
    "Gap Analysis": {
        "Minimum Gap Diameter": 10
    }
}

# Save parameters to a YAML file
param_file = "cabana_params.yaml"
with open(param_file, 'w') as f:
    yaml.dump(params, f)

print(f"Parameter file created: {param_file}")
```

### 4. Creating Output Directory
Next, let's create an output directory where the results will be saved.

```python
# Create output directory
output_dir = "./cabana_results"
os.makedirs(output_dir, exist_ok=True)
print(f"Output directory created: {output_dir}")
```

### 5. Basic Usage
Now, let's use Cabana to analyze a single image.

```python

# Specify the path to your input image
input_image_path = "/Users/lxfhfut/Dropbox/Garvan/Cabana/Input/540.vsi - 20x_BF multi-band_01Annotation (Ellipse) (Tumor)_0.tif"  # Replace with your image path

# Create a CabanaIndividual instance
analyzer = Cabana(
    param_file=param_file,
    input_image_path=input_image_path,
    out_folder=output_dir,
    ignore_large=True  # Set to False if you want to split large images into smaller blocks
)

# Run the analysis
success = analyzer.run()

if success:
    print("Analysis completed successfully!")
else:
    print("Analysis failed or was skipped.")
```

### 8. Batch Processing
There is a batch processer that can be used to process multiple images in a loop with checkpoint re. Here's an example of how to process all images in a directory.

```python
import time
from log import Log
from batch_processor import BatchProcessor

start_time = time.time()
batch_processor = BatchProcessor(5)
batch_processor.run()
time_secs = time.time() - start_time
hours = time_secs // 3600
minutes = (time_secs % 3600) // 60
seconds = (time_secs % 3600) % 60
Log.logger.info("--- {:.0f} hours {:.0f} mins {:.0f} seconds ---".format(hours, minutes, seconds))
```

Alternatively, you can use Cabana to process all images in a folder individually

```python
import glob
from tqdm import tqdm

# Directory containing images
image_dir = "/Users/lxfhfut/Dropbox/Cabana/Input"  # Replace with your image directory

# Get all image files
image_paths = glob.glob(os.path.join(image_dir, "*.jpg")) + \
              glob.glob(os.path.join(image_dir, "*.png")) + \
              glob.glob(os.path.join(image_dir, "*.tif"))

print(f"Found {len(image_paths)} images")

# Create a dataframe to store aggregate statistics
all_stats = pd.DataFrame()

# Process each image
for image_path in image_paths:
    img_name = os.path.basename(image_path)
#     print(f"Processing {img_name}...")
    
    # Create output directory for this image
    img_output_dir = os.path.join(output_dir, os.path.splitext(img_name)[0])
    os.makedirs(img_output_dir, exist_ok=True)
    
    # Create analyzer and run
    analyzer = Cabana(
        param_file=param_file,
        input_image_path=image_path,
        out_folder=img_output_dir,
        ignore_large=True
    )
    
    success = analyzer.run()
    
    if success:
        # Add image name to stats and append to all_stats
        analyzer.stats['Image_Path'] = image_path
        all_stats = pd.concat([all_stats, analyzer.stats], ignore_index=True)
    else:
        print(f"Failed to analyze {img_name}")
    
    print("\n" + "-"*50 + "\n")

# Save aggregate statistics
if not all_stats.empty:
    all_stats.to_csv(os.path.join(output_dir, "all_images_stats.csv"), index=False)
    print(f"Aggregate statistics saved to {os.path.join(output_dir, 'all_images_stats.csv')}")
else:
    print("No images were successfully analyzed.")
```

#### 8.2 Correlation Analysis
Let's look at correlations between different metrics.

```python
# Calculate correlation matrix
if not all_stats.empty:
    # Select numeric columns for correlation analysis
    numeric_cols = all_stats.select_dtypes(include=['number']).columns
    
    # Calculate correlation matrix
    corr_matrix = all_stats[numeric_cols].corr()
    
    # Plot correlation heatmap
    plt.figure(figsize=(14, 12))
    plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
    plt.colorbar(label='Correlation coefficient')
    plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=90)
    plt.yticks(range(len(numeric_cols)), numeric_cols)
    plt.title('Correlation Matrix of Fiber Metrics')
    
    plt.tight_layout()
    plt.show()
    
    # Identify most strongly correlated metrics
    print("Top 10 strongest correlations:")
    # Flatten the correlation matrix, remove diagonal (self-correlations)
    corr_pairs = []
    for i in range(len(numeric_cols)):
        for j in range(i+1, len(numeric_cols)):
            corr_pairs.append((numeric_cols[i], numeric_cols[j], corr_matrix.iloc[i, j]))
    
    # Sort by absolute correlation value
    corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)
    
    # Print top 10
    for i, (col1, col2, corr_val) in enumerate(corr_pairs[:10]):
        print(f"{i+1}. {col1} and {col2}: {corr_val:.3f}")
```
